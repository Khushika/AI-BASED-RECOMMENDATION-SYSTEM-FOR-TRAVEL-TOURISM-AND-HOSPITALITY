```python
# Step 1: Data Preprocessing
def preprocess_data(hotel_data):
    # Handle missing values
    hotel_data.dropna(inplace=True)
    # Clean text data
    hotel_data['hotel_description'].fillna('', inplace=True)
    hotel_data['hotel_facilities'].fillna('', inplace=True)
    # Combine text features for NLP
    hotel_data['text_features'] = hotel_data['hotel_brand'] + ' ' +hotel_data['room_type']+ ' ' + hotel_data['hotel_description'] + ' ' + hotel_data['hotel_facilities']
    return hotel_data
```

```python
# Step 2: NLP-Based Keyword Search
def keyword_search(user_location, tfidf_matrix, tfidf_vectorizer, hotel_data, selected_categories=None):
    # Calculate distances from user location to hotel locations
    hotel_data['distance_to_user_km'] = hotel_data.apply(lambda row: geodesic(user_location, (row['latitude'], row['longitude'])).kilometers, axis=1)

    # Get user state
    geolocator = Nominatim(user_agent="hotel_locator")
    user_location_info = geolocator.reverse(user_location)
    user_state = user_location_info.raw['address']['state']

    # Filter places based on state
    recommended_places = hotel_data[hotel_data['state'] == user_state]

    # Sort places based on distance from user location
    recommended_places = recommended_places.sort_values(by='distance_to_user_km')

    return recommended_places
```

```python
 #Step 3: Map Visualization with Color-Coded Ratings
def display_recommendations_on_map(recommendations, user_location):
    # Create a map centered around the search location
    map_places = folium.Map(location=user_location, zoom_start=12)

    # Add avatar marker for user location with a colorful icon
    folium.Marker(location=user_location, popup="Your Location", icon=folium.Icon(color='green', icon='user')).add_to(map_places)

    # Create MarkerClusters for different rating groups
    ratings = sorted(recommendations['hotel_star_rating'].unique(), reverse=True)
    for rating in ratings:
        rating_group = recommendations[recommendations['hotel_star_rating'] == rating]
        marker_cluster = MarkerCluster().add_to(map_places)

        # Add markers for each place in the rating group
        for index, row in rating_group.iterrows():
            place_location = (row['latitude'], row['longitude'])
            popup_html = f"<b>{row['hotel_brand']}</b><br><i>Address:</i> {row['address']}<br><i>Distance to Your Location:</i> {row['distance_to_user_km']:.2f} km"
            popup_html += f"<br><i>Rating:</i> {row['hotel_star_rating']}"
            folium.Marker(location=place_location, popup=popup_html).add_to(marker_cluster)

    # Display the map
    display(map_places)
```

```python
from IPython.display import display, HTML
import ipywidgets as widgets
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import folium
from folium.plugins import MarkerCluster
import pandas as pd

# Load Goibibo Indian hotel dataset
hotel_data = fdata

# Step 1: Data Preprocessing
def preprocess_data(hotel_data):
    # Handle missing values
    hotel_data.dropna(inplace=True)
    # Clean text data
    hotel_data['hotel_description'].fillna('', inplace=True)
    hotel_data['hotel_facilities'].fillna('', inplace=True)
    # Combine text features for NLP
    hotel_data['text_features'] = hotel_data['hotel_brand'] + ' ' + hotel_data['hotel_description'] + ' ' + hotel_data['hotel_facilities']
    return hotel_data

# Step 2: NLP-Based Keyword Search
def keyword_search(user_location, tfidf_matrix, tfidf_vectorizer, hotel_data, selected_categories):
    # Calculate distances from user location to hotel locations
    hotel_data['distance_to_user_km'] = hotel_data.apply(lambda row: geodesic(user_location, (row['latitude'], row['longitude'])).kilometers, axis=1)

    # Get user state
    geolocator = Nominatim(user_agent="hotel_locator")
    user_location_info = geolocator.reverse(user_location)
    user_state = user_location_info.raw['address']['state']

    # Filter places based on state
    recommended_places = hotel_data[hotel_data['state'] == user_state]

    # Filter places based on selected categories and their values
    for category, value in selected_categories.items():
        if value == 'low':
            recommended_places = recommended_places[recommended_places[category] < recommended_places[category].quantile(0.33)]
        elif value == 'medium':
            recommended_places = recommended_places[(recommended_places[category] >= recommended_places[category].quantile(0.33)) & (recommended_places[category] < recommended_places[category].quantile(0.66))]
        elif value == 'high':
            recommended_places = recommended_places[recommended_places[category] >= recommended_places[category].quantile(0.66)]
        elif value == 'more':
            recommended_places = recommended_places[recommended_places[category] > recommended_places[category].mean()]
        elif value == 'less':
            recommended_places = recommended_places[recommended_places[category] < recommended_places[category].mean()]

    # Sort places based on distance from user location
    recommended_places = recommended_places.sort_values(by='distance_to_user_km')

    return recommended_places

# Step 3: Display Hotel Details
def display_hotel_details(hotel, selected_categories):
    details = f"<b>Hotel Name:</b> {hotel['hotel_brand']}<br>"
    for category in selected_categories:
        if category in hotel:
            details += f"<b>{category.capitalize()}:</b> {hotel[category]}<br>"
    display(HTML(details))
    print()

# Step 4: Usage Example
def main():
    # Preprocess the data
    hotel_data_processed = preprocess_data(hotel_data)

    # Create text input widget for user location
    location_input = widgets.Text(description="User Location:")

    # Define columns to exclude from the dropdown
    exclude_columns = ['crawl_date', 'image_count', 'page_url', 'qts', 'query_time_stamp', 'review_count_by_category']

    # Create multi-select dropdown widget for selecting categories with values
    categories_with_values = {col: ['low', 'medium', 'high', 'more', 'less'] for col in hotel_data_processed.columns if col not in exclude_columns}
    categories_dropdown = widgets.SelectMultiple(
        options=categories_with_values.keys(),
        description='Select Categories:',
        style={'description_width': 'initial'}
    )

    # Create search button
    search_button = widgets.Button(description="Search")

    # Define function to handle search button click
    def handle_search_button_click(b, hotel_data_processed):
        user_location_str = location_input.value

        # Geocode user location
        geolocator = Nominatim(user_agent="hotel_locator")
        user_location = geolocator.geocode(user_location_str)
        if user_location is None:
            print("Invalid user location. Please try again.")
            return
        user_coordinates = (user_location.latitude, user_location.longitude)

        # Get selected categories and their values
        selected_categories = {category: categories_with_values[category] for category in categories_dropdown.value}

        # Perform keyword search
        recommendations = keyword_search(user_coordinates, None, None, hotel_data_processed, selected_categories)

        # Display hotel details for each recommendation
        for _, hotel in recommendations.iterrows():
            display_hotel_details(hotel, selected_categories)

    # Attach click event handler to search button
    search_button.on_click(lambda b: handle_search_button_click(b, hotel_data_processed))

    # Display widgets
    display(location_input, categories_dropdown, search_button)

if __name__ == "__main__":
    main()
```

```python

import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import word_tokenize
from IPython.display import HTML, display
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob

# Load the dataset
data = pd.read_csv(r"C:\Users\mamid\Downloads\HOTEL REVIEWS DATASET.csv")

# Preprocess the dataset (cleaning, handling missing values, etc.)
data['REVIEW'].fillna('', inplace=True)

# Tokenize user reviews
data['TOKENIZED_REVIEW'] = data['REVIEW'].apply(lambda x: word_tokenize(str(x).lower()))

# Combine tokenized reviews into strings
data['CLEANED_REVIEW'] = data['TOKENIZED_REVIEW'].apply(lambda x: ' '.join(x))

# Vectorize the reviews using TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(data['CLEANED_REVIEW'])

# Dictionary to store user search history
user_search_history = {}

# Function to interpret user search query and recommend hotels
def recommend_hotels(user_id, user_query, data, vectorizer, tfidf_matrix):
    # Store user search history
    if user_id not in user_search_history:
        user_search_history[user_id] = []
    user_search_history[user_id].append(user_query)

    # Vectorize the user query using TF-IDF
    user_query_vector = vectorizer.transform([user_query])

    # Calculate similarity between user query and hotels
    similarity_scores = cosine_similarity(user_query_vector, tfidf_matrix)

    # Get the indices of hotels with similar keywords in reviews as the search query
    similar_hotel_indices = similarity_scores.argsort()[0][-5:][::-1]

    # Retrieve the hotel details for unique hotels
    unique_hotels = set()
    recommended_hotels = []
    for index in similar_hotel_indices:
        hotel_name = data.iloc[index]['1 NAME']
        if hotel_name not in unique_hotels:
            unique_hotels.add(hotel_name)
            rating = data.iloc[index]['RATING']
            price = data.iloc[index]['PRICE']
            recommended_hotels.append((hotel_name, rating, price))

    return recommended_hotels

# Collaborative filtering function to predict user ratings for hotels
def collaborative_filtering(user_id, user_query, data):
    # Get search history of similar users
    similar_user_search_history = []
    for history_user_id, search_history in user_search_history.items():
        if history_user_id != user_id:
            similar_user_search_history.extend(search_history)

    if similar_user_search_history:
        # Filter the dataset based on similar user search history
        filtered_data = data[data['CLEANED_REVIEW'].isin(similar_user_search_history)]

        # Get average rating for the filtered hotels
        average_rating = filtered_data['RATING'].mean()
    else:
        # If no similar user search history available, return a default value or handle appropriately
        average_rating = data['RATING'].mean()  # Default to overall average rating of all hotels

    return average_rating

# Sentiment analysis function
def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Function to generate HTML for displaying reviews
def generate_review_html(hotel_reviews):
    reviews_html = ""
    for sentiment in ['positive', 'negative', 'neutral']:
        sentiment_reviews = hotel_reviews[hotel_reviews['SENTIMENT'] == sentiment]['REVIEW'].tolist()
        if sentiment_reviews:
            reviews_html += f"<h2>{sentiment.capitalize()} Reviews</h2>"
            reviews_html += "<ul>"
            for review in sentiment_reviews[:3]:  # Display only top 3 reviews
                reviews_html += f"<li>{review}</li>"
            reviews_html += "</ul>"
    return reviews_html

# Function to display reviews for a particular hotel when review button is clicked
def display_reviews(hotel_name, user_query, data):
    hotel_reviews = data[data['1 NAME'] == hotel_name].copy()  # Copy to avoid SettingWithCopyWarning
    hotel_reviews['SENTIMENT'] = hotel_reviews['REVIEW'].apply(analyze_sentiment)
    reviews_html = generate_review_html(hotel_reviews)
    display(HTML(f"<h3>Reviews for {hotel_name}</h3>"
                 f"<p><strong>Search Query:</strong> {user_query}</p>"
                 f"{reviews_html}"))

# Get dynamic inputs from the user
user_id = input("Enter user ID: ")
user_query = input("Enter your search query: ")

# Recommend hotels based on the user query (Content-Based Filtering)
recommended_hotels_content_based = recommend_hotels(user_id, user_query, data, vectorizer, tfidf_matrix)

# Display unique recommended hotels along with their ratings, price, and review buttons (Content-Based Filtering)
print("Recommended Hotels based on search query (Content-Based Filtering):", user_query)
for hotel_name, rating, price in recommended_hotels_content_based:
    display(HTML(f"<h1>{hotel_name}</h1>"
                 f"<p><strong>Rating:</strong> {rating}</p>"
                 f"<p><strong>Price:</strong> {price}</p>"))
    display_reviews(hotel_name, user_query, data)

# Predict user rating based on collaborative filtering
predicted_rating = collaborative_filtering(user_id, user_query, data)
if not pd.isna(predicted_rating):
    print("Predicted Rating based on similar users' search history:", predicted_rating)
else:
    print("No similar users' search history available to predict the rating.")
```
